## Ejemplos de Mecanico

```python
# Lista para guardar los nombres de los mecánicos
mecanicos = []

def agregar_mecanico():
    nombre = input("Nombre del mecánico: ").strip()
    if nombre:
        mecanicos.append(nombre)
        print(f"🔧 {nombre} ha sido dado de alta en el taller.")
    else:
        print("⚠️ El nombre no puede estar vacío.")

def listar_mecanicos():
    if not mecanicos:
        print("📭 El taller está vacío, no hay mecánicos registrados.")
    else:
        print("\n--- LISTA DE MECÁNICOS EN TURNO ---")
        for i, nombre in enumerate(mecanicos, 1):
            print(f"{i}. {nombre}")

def buscar_mecanico():
    nombre_buscar = input("Nombre del mecánico a buscar: ").strip()
    if nombre_buscar in mecanicos:
        print(f"✅ Confirmado: {nombre_buscar} se encuentra en las instalaciones.")
    else:
        print(f"❌ No se encontró a ningún mecánico llamado '{nombre_buscar}'.")

def modificar_mecanico():
    nombre_buscar = input("Nombre del mecánico a cambiar: ").strip()

    if nombre_buscar in mecanicos:
        # Buscamos la posición en la lista para reemplazarlo
        indice = mecanicos.index(nombre_buscar)
        nuevo_nombre = input("Ingrese el nuevo nombre: ").strip()
        mecanicos[indice] = nuevo_nombre
        print(f"⚙️ Cambio realizado: '{nombre_buscar}' ahora es '{nuevo_nombre}'.")
    else:
        print("❌ Ese nombre no figura en nuestros registros.")

# --- LÓGICA CON CICLO FOR ---
# El ciclo se ejecutará hasta 50 veces a menos que se use la opción 'Salir'
for intento in range(50):
    print("\n--- GESTIÓN DE PERSONAL (MECÁNICOS) ---")
    print("1. Registrar")
    print("2. Ver todos")
    print("3. Localizar")
    print("4. Editar nombre")
    print("5. Salir")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        agregar_mecanico()
    elif opcion == "2":
        listar_mecanicos()
    elif opcion == "3":
        buscar_mecanico()
    elif opcion == "4":
        modificar_mecanico()
    elif opcion == "5":
        print("Saliendo del sistema... ¡Ten un buen día!")
        break
    else:
        print("⚠️ Opción no válida. Intente de nuevo.")
print("\nPrograma realizado por Meredith Aguirre NL = 2")
```
