## Ejemplos de Mecanico

```python
# Diccionario global para almacenar los datos
# Estructura: { id: {'nombre': '...', 'esp': '...'} }
mecanicos = {}

def crear_mecanico(nombre, esp):
    """Retorna un diccionario con la estructura del mecánico."""
    return {"nombre": nombre, "esp": esp}

def formatear_mecanico(idx, datos):
    """Devuelve una cadena legible con la información."""
    return f"ID: {idx} | Nombre: {datos['nombre']} | Especialidad: {datos['esp']}"

def gestionar(op):
    if op == '1':
        idx = input("ID: ")
        nombre = input("Nombre: ")
        esp = input("Especialidad: ")
        # Guardamos un diccionario en lugar de una instancia de clase
        mecanicos[idx] = crear_mecanico(nombre, esp)
        print(">>> Agregado.")

    elif op == '2':
        if not mecanicos:
            print("No hay mecánicos registrados.")
        for idx, datos in mecanicos.items():
            print(formatear_mecanico(idx, datos))

    elif op in ['3', '4', '5']:
        idx = input("Ingrese ID: ")
        if idx in mecanicos:
            if op == '3':
                print(formatear_mecanico(idx, mecanicos[idx]))

            elif op == '4':
                mecanicos[idx]['nombre'] = input("Nuevo nombre: ")
                mecanicos[idx]['esp'] = input("Nueva esp: ")
                print(">>> Modificado.")

            elif op == '5':
                del mecanicos[idx]
                print(">>> Eliminado.")
        else:
            print("No encontrado.")

# --- Bucle Principal ---
while True:
    print("\n1.Agregar 2.Listar 3.Buscar 4.Modificar 5.Borrar 6.Salir")
    op = input("Opción: ")
    if op == '6':
        break
    gestionar(op)

print("\nPrograma realizado por Meredith Aguirre NL = 2")
```
